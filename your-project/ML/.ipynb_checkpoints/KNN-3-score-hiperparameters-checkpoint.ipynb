# Let's try to find a better random seed.

```python
X = df.drop(["risk_group"], axis=1)
y = df["risk_group"]

scaler = MinMaxScaler()
X = scaler.fit_transform(X)

#from 0 to 300 -> results (i just saw a 74 in seed 57, 74.1 at 116, 74.04 at 135)
score_list = []

for i in range(301):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i)

    model = KNeighborsClassifier(n_neighbors = 29)
    clf = model.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    score_list.append(accuracy_score(y_pred, y_test))
    print('iteration number',i,':', accuracy_score(y_pred, y_test))
```
